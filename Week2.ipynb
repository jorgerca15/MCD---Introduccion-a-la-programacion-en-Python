## **Listas**

### Tema 1: Definicion de listas y acceso a sus elementos

- Una lista guarda una secuencia de elementos
- Un elemento en cada casillero
- Cada casillero puede guardar un distinto tipo de elemento
- Listas son mutables: podemos cambiarlos cuando necesitamos
- Pueden crecer segun necesidades del programa mediante append e insert
- Elementos puedes ser eliminados mediante remove y pop

Imaginemos que queremos guardar la temperatura de la semana

```python
temp = [30.5,19.8,12.3,31.0,19.9,34.4,25.3]
print(temp[0])
print(temp[3:5])
```

Las listas se pueden modificar:

```python
temp[2] = 22.3
print(temp)
```

### Tema 2: Como recorrer una lista

```python
print(len(temp))
```

Recorrer una lisat con la funcion for in range: Modo 1

```python
for i in range(len(temp)):
    print(temp[i])
```

Recorrer una lisat con la funcion for range: Modo 2

```python
for elem in temp:
    print(elem)
```

### Tema 5: Metodos adicionales de listas

```python
# Metodo index:

temp = [30.5, 19.8, 22.3, 31.0, 16.5, 34.4, 25.3]
dias = ['Lu','Ma','Mi','Ju','Vi','Sa','Do'] # Creamos la lista dias
# print(dias)
temperatura = 22.3

if temperatura in temp:
    ind = temp.index(temperatura)
    print(dias[ind])
```

## **Tuplas**

### Tema 1 : que es una tupla

- Coleccion ordenada e inmutable de elementos del mismo o diferente tipo 
- Al igual que con las listas, para la selecciÃ³n se usan corchetes []





### Tema 2: Diferencia entre listas y tuplas

```python
d = ('Lu','Ma','Mi','Ju','Vi','Sa','Do') # Creamos la tupla dias
print(d[3])
```

- listas [] -> mutables (si se pueden modifiar)

- tuplas() -> inmutables (no pueden modificarse)

### Tema 3: Creacion de tuplas

#### **Nota:** Los parentesis se usan usualmente, pero no son obligatorios

### Tema 4: Manejo de tuplas

```python
m1 = ('Toy Story', 1995, '01:21',['animacion','comedia']) #creando la tupla
(nombre, anno, duracion, categorias) = m1 

print(anno)
print(categorias)
```

Esto se usa para ordenar listas



### Tema 5: Retorno de una funcion

```python
# Creacion de una funcion para convertir las tazas de harina

def convertir(taza):
    g = taza*1.25
    o = taza*4.41
    l = taza*0.28

    return g,o,l # Retornar 3 valores separados por coma como una tupla

# Evaluando los valores

gramos, onzas, libras = convertir(1.5)
print(gramos)



print(convertir(1.5))
```

```python
# Pregunta 2

lista = [-40, -1, 2, 5, 16, 72, 100]
print(lista)
```

```python
def agregar_entero(lista, entero):
    for i in range(len(lista)):
        if entero < lista[i]:
            lista.insert(i,entero)
            break
        elif entero > lista[len(lista)-1]:
            lista.insert(len(lista),entero)
            break
    return lista

agregar_entero(lista,99)
print(lista)
```
