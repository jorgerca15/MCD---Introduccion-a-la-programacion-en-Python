# Como guardar informacion multidimensional en listas de listas

### Ejemplo 1

```python
informacion = "Nombre,Edad,Activo\nAlicia,34,True\nBernardo,40,False\nCarolina,39,False"
print(informacion)

#Queremos una lista de listas:

lista = informacion.split("\n")
#print(lista)

for i in range(len(lista)):
    lista[i] = lista[i].split(",")
    
print(lista)
```

```python
for lista_int in lista[1:]:
    #print(lista_int)
    lista_int[1] = int(lista_int[1])
    lista_int[2] = (lista_int[2] == "True")



print(lista)
```
